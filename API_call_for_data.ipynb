```python
import requests
import csv
import datetime

# URL of the CSV file
url = "https://data.boston.gov/dataset/7ae458ed-cd6a-47fa-8571-e323e488b70b/resource/dd657c02-3443-4c00-8b29-56a40cfe7ee4/download/tmpk9ahbct2.csv"

# File name of the downloaded CSV file
filename = "data_" + datetime.date.today().strftime("%Y-%m-%d") + ".csv"

# Send HTTP GET request to download CSV file
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse CSV data and write to file
    csv_data = response.content.decode('utf-8')
    with open(filename, 'w') as f:
        f.write(csv_data)
    print(f"File {filename} downloaded successfully.")
else:
    print("Failed to download CSV file.")
```
